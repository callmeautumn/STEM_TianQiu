Sort the collection alphabetically by song name

````python
# song names array
song_names = this_is_my_jam_dataset[:, 2]


# Convert strings to ASCII encoding
ascii_encoded_data = []
for item in song_names:
    item = item.encode('ascii', 'ignore')
    ascii_encoded_data.append(item)

# Convert the byte sequence back to a string
ascii_string_data = [item.decode('ascii') for item in ascii_encoded_data]

#Convert the data to a set to remove duplicate items
unique_data = list(set(ascii_string_data))

data = [item for item in unique_data if item]

# Rearrange the original data based on the sorted indices
sorted_data = np.sort(data)

# Remove quotation marks and spaces
sorted_data1 = [item.strip('"" ').strip() for item in sorted_data]

sorted_data1 = [item.encode('ascii', 'ignore').decode() for item in data]

# Use numpy.char.lower() to convert strings to lowercase for sorting
sorted_indices = np.argsort(np.char.lower(sorted_data1))
sorted_data2 = np.array(data)[sorted_indices]


print(sorted_data2)
````

Given this sorted list, print 10 last tracks

````python
#Your code here
last_ten_data = sorted_data2[-10:]
print(last_ten_data)
````
